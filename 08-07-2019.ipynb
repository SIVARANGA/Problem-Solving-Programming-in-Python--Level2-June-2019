```python
# Function to generate all Perfect numbers in a given range

def perfect(n):
    sum1=0
    for i in range(1,n):
        if n%i==0:
            sum1=sum1+i
    if sum1==n:
        return True
    else:
        return False
        
perfect(6)

def NPrineNumbers(lb,up):
    for num in range(lb,up+1):
        if perfect(num)==True:
            print(num)
            
NPrineNumbers(30,50)
```

```python
# Function to calculate the average of all factorials in a given range

def AverageFactorial(lb,up):
    count=0
    sum1=0
    fact=1
    for num in range(lb,up+1):
        for i in range(1,num):
            fact=fact*i
            #print("fact",fact)
        count=count+1
        sum1=sum1+fact
        #print(count,"count")
        #print(sum1,"sum")
    print(sum1,"s")
    print(count,"c")
    average=sum1/count
    print(average)
    
AverageFactorial(1,4)
```

```python
# Function to generate N odd Armstrong numbers

def ArmStrongNumber(n):
    m=n
    d=len(str(n))
    sum1=0
    while(n>0):
        mod=n%10
        p=mod**d
        sum1=sum1+p
        n=(n//10)
        
    if sum1==m:
        return True
    else:
        return False
    
ArmStrongNumber(456)

def checkoddarmstrongNumber(lp,up):
    for num in range(lp,up+1):
        if ArmStrongNumber(num)==True:
            if num%2!=0:
                print(num)

checkoddarmstrongNumber(340,380)
```

```python
# Function to generate Multiplication table for a number in a given range

def MultificationTable(num,lp,up):
    for i in range(lp,up+1):
        op=num*i
        print(num,"*",i,"=",op)
        
MultificationTable(10,5,10)
```
