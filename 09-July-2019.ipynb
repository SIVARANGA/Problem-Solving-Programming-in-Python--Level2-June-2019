```python
# Function to calculate the average first N Prime numbers

def isPrime(n):
    count=0
    for i in range(1,n+1):
        if n%i==0:
            count+=1
    if count==2:
        return True
    else:
        return False
    
#isPrime(11)

def AveragePrimeNumbers(m):
    num=2
    count1=0
    
    sum1=0
    while(count1<m):
        #print(isPrime(num))
        if isPrime(num)==True:
            count1+=1
            #print(count1,"c")
            sum1=sum1+num
            #print(sum1)
            num=num+1 
            if count1==m:
                break
    average=sum1/count1
    print(average)
    
AveragePrimeNumbers(3)
```
