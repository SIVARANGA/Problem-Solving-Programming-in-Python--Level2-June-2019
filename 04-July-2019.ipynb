```python
# Linear graph

import matplotlib.pyplot as plt

plt.plot([1,2,3],[1,2,3])

plt.title("Basic linear graph")

plt.show
```

```python
# Graph to plot distance and time

distance=list(range(2,11,2))
time=list(range(1,6))

plt.title("Distance vs time")

plt.grid()
plt.plot(time,distance,'b*')

plt.xlabel("Time ")
plt.ylabel("Distance")

plt.show

```

```python
# Velocity and accerlation graph v=a*t

t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]

plt.plot(t,v,'g*')
plt.grid()
plt.show()
```

```python
n=list(range(10))

k=[i**4 for i in (n)]

plt.plot(n,k)
plt.show()
```

```python
#to print a sine wave
import numpy as np
y=np.arange(0,4*np.pi,0.2)
x=np.sin(y)


plt.grid()
plt.plot(y,x,"g*")

plt.savefig("Images/sine.png")
```
