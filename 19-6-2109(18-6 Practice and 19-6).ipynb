```python
details={}

def addContacts(name,phonenumber):
    details[name]=phonenumber
    return details

addContacts("Ranga",9640494680)

addContacts("Ranga1",9640494680)

details
```

```python
details={}

def addcontact(name,phonenumber):
    if name not in details:
        details[name]=phonenumber
        return details
    else:
        print("%s already exist",name)
    
addcontact("Alekhya",4655)

addcontact("Alekhya1",465555)
addcontact("Ranga",465555)

details
```

```python
def serachContac(name):
    if name in details:
        print(name,":",details[name])
    else:
        print("%s is not in dictionary",name)
        
serachContac("Alekhya")
```

```python
def modify(name,phonenumber):
    if name in details:
        details[name]=phonenumber
        return details
    else:
        print("%s is not availble ",name)
        
modify("Alekhya","688888")
```

```python
def delete(name):
    if name in details:
        details.pop(name)
        return details
    else:
        print("%s no name available",name)
        
delete("Alekhya")
        
```

```python
def printdetails():
    for keys,value in details.items():
        print(keys,":",value)
        
        
printdetails()
    
```

```python
def importContacts(newcontacts):
    details.update(newcontacts)
    print(len(newcontacts),"add in details")

newcontacts={"Siva":"98","Rsaha":"87"}
importContacts(newcontacts)
```
