```python
def TablePrint(table,start,end):
    for value in range(start,end+1):
        op=table*value
        print(table,'X',value,"=",op)
        
TablePrint(10,10,20)
```

```python
#Function to search for data in a list

def searchElement(li,value):
    for i in range(0,len(li)):
        if li[i]==value:
            return i
    return -1
      
searchElement(li,7)
```

```python
#Count the occurances of a character in a list
def CountSameNumberInList(string,chara):
    count=0
    for i in string:
        if chara==i:
            count=count+1
    print(count)
    
    
CountSameNumberInList("Ranga","n")
    
```
