# Functions

```python
# Reverse the string using function

def reversestring(string):
    a=string[-1: :-1]
    print(a)
    
reversestring("Ranga")
```

```python
def reverse(s1):
    return s1[-1::-1]
```

```python
reverse("Hii how are you")
```

```python
#function to check wheather the given string is palindrome or not 

def palindrome(s1):
    a=s1[::-1]
    if s1==a:
        return True
    else:
        return False
    
palindrome("RangnaR")
```

```python
#Greatest number in 4  numbers

def greatest(n1,n2,n3,n4):
    if n1>n2 and n1>n3 and n1>n4:
        return "n1"
    elif n2>n3 and n2>n3:
        return "n2"
    elif n3>n4:
        return "n3"
    else:
        return "n4"
greatest(2,4,12,12)
```

```python
def printNaturalNumber(n):
    for counter in range(1,n+1):
        print(counter,end=" ")
        
printNaturalNumber(123)
```

```python
#Function to print all numbers divisible by 6 and not a factor of 100 in a given range(lb,ub) inclusive

def range1(b,u):
    for i in range(b,u+1):
        if 6%i==0 and 100%i!=0:
            return(i)
        
range1(1,10)
            
```

```python
# Function to find the avearge of cubes of all even numbers in a given range 

def average(a,b):
    s=0
    op=0
    for i in range(a,b+1):
        if i%2==0:
            op=i*i*i
            s=s+op
    print(s)

average(1,10)
```
