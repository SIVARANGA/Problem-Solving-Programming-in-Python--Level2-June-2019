### Map



```python
def power(a,n):
    h=a**n
    return h
li=[1,2,3,1]
li1=[2,4,6,2]
list(set(map(power,li,li1)))
```

```python
li1=['2','4','6','2','a','e']
def isDigit(n):
    n=str(n)
    
    if n.isdigit():
        return 0
    return 1

isDigit(li1)

list(filter(isDigit,li1))
```

```python
# Prime or not
n=list(range(1,11))
def primeNumber(n):
    count=0
    for i in range(1,n+1):
        if n%i==0:
            count=count+1
    if count==2:
        return i
    return 0
    
    
list(map(primeNumber,n))
```

```python
p1=[num for num in range(int(input()),int(input())+1) if primeNumber(num)]
p1
```

```python
list(map(lambda x:x**2 ,[1,2,3,4,4]))
```

```python
from random import randint

ex1=[randint(1,25) for i in range(10)]
ex2=[randint(1,25) for j in range(10)]

p=list(map(lambda x,y: (x+y)//2 ,ex1,ex2))

pa=list(filter(lambda h: h>10,p))
pa1=list(filter(lambda k:k<=10,p))
pa1
```

## Data Encryption

Encryption based on Key



```python
keypath='Datafiles/key.txt'
def generateKey(keypath):
    with open(keypath,'w') as f:
        for i in range(10):
            if i<6:
                f.write(str(i)+' '+str(i+4)+"\n")
            else:
                f.write(str(i)+' '+str(i-6)+"\n")
                
generateKey(keypath)
    
```

```python
keyfile='Datafiles/key.txt'
def dictor(keyfile):
    key={}
    with open(keyfile,'r') as f:
        for line in f:
            line=line.split()
            key[line[0]]=line[1]
    return key


dictor(keyfile)
```
