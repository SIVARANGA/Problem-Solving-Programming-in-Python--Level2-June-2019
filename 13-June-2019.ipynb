```python
#Function to print alternative numbers

def alternativeNum(start,end):
    for value in range(start,end+1,10):
        print(value,end=" ")
    return
        
alternativeNum(100,120)
```

```python
#Function to print reverse of given range in the single line

def printreverse(high,low):
    for value in range(high,low-1,-1):
        print(value,end=" ")
        
printreverse(30,20)
```

```python
#Function to print the odd numbers in a given range
def printodd(high,low):
    for va in range(high,low-1,-1):
        if va%2==1:
            print(va,end=" ")
            
printodd(200,140)
```

```python
#Function to calculate the sum of numbers in a range

def sumofnumbers(start,end):
    su=0
    for value in range(start,end+1):
        su=su+value
    return su
      
sumofnumbers(1,100)
```

```python
#Function to calculate the average of a numbers in a given range

def averageinrange(start,end):
    su=0
    for value in range(start,end+1):
        su=su+value
        
    avg=su/(end+1-start)
    return avg

averageinrange(1,10)
```

```python
def isPrime(n):
    flag = True
    for i in range(2, n//2+1):
        if n % i == 0:
            flag = False
            return flag
    return flag

isPrime(4)
```

```python
def avgNPrimes(n):
    primeCount = 0
    sum = 0
    seqCount = 2
    while(primeCount < n):
        if isPrime(seqCount):
            primeCount += 1
            sum += seqCount
            #print(seqCount)
        seqCount += 1
    return sum/n

avgNPrimes(10)
```
