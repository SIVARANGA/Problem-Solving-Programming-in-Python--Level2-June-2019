```python
# Write a Python function that takes a list of words and returns the length of the longest one

def find_largest_word(words_list):
    words_len=[]
    for n in words_list:
        words_list.append((len(n),n))
        
    words_list.sort()
    return words_list[-1][1]

find_largest_word(['ranga','siva','nayakulu'])
        
```

```python
def find_longest_word(words_list):
    word_len = []
    for n in words_list:
        word_len.append((len(n), n))
        #print(word_len)
    word_len.sort()
    return word_len[-1][1]

print(find_longest_word(["PHP", "Exercises", "Backend"]))
```

```python
#Write a Python function to convert a given string to all uppercase if it contains at least 2 uppercase characters 
#in the first 4 characters

def upperconvert(string):
    c=0
    for char in string[:4]:
        if char.isupper():
            c=c+1
    if c>=2:
        print(string.upper())
        
    else:
        print(string)
        
k=input("Enter the string")
upperconvert(k)
```
