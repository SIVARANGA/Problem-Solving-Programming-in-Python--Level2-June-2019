```python
def prime(number):
    a=0
    for n in range(1,number+1):
        if number%n==0:
            a+=1
    if a==2:
        return True
        
        
num=int(input())
for times in range(num):
    n=int(input())
    string=input()
    s=""
    for char in string:
        m=ord(char)
        #while True:
        for g in range(m,m+20):
            if prime(g)==True:
                p1=g
                break
                
        #while True:
        for h in range(m,m-20,-1):
            if prime(h)==True:
                p2=h
                break
                
        print(p1,"positive")
        print(p2,"negative")
        
        if  abs(m-p1)>=abs(m-p2):
            s+=chr(p2)
        elif abs(m-p1)<abs(m-p2):
            s+=chr(p1)
            
    print(s)
```
