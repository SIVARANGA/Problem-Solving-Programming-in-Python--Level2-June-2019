```python
# Function to print all numbers divisible by 6 and not a factor of 100 in a given range(lb, ub) inclusive

def divisiblebysix(lb,up):
    for i in range(lb,up+1):
        if i%6==0 and 100%i!=0:
            print(i)
            
            
divisiblebysix(int(input()),int(input()))
```

```python
# Function to find the average of cubes of all even numbers in a given range(lb, ub) inclusive

def AveargecubesEvenNumber(lb,up):
    count=0
    cube=0
    for num in range(lb,up+1):
        if num%2==0:
            count+=1
            cube=num**3+cube
            
    average=cube/count
    print(average)
AveargecubesEvenNumber(1,10)

```

```python
# Function to generate the list of factors for a given number

def factors(num):
    for i in range(1,num+1):
        if num%i==0:
            print(i)
                       
factors(6)
```

```python
# Function to calculate the factorial of a given number

def factorial(num):
    fact=1
    for i in range(1,num+1):
        fact=fact*i
    print(fact)
    
factorial(int(input()))
        
```

```python
#Function to check if a given number is Prime

def PrimeNumber(num):
    count=0
    for i in range(1,num+1):
        if num%i==0:
            count=count+1
            
            
    if count==2:
        return 1
    else:
        return -1
        
PrimeNumber(int(input()))

#Another logic

'''def primenumber1(n):
    for i in range(2,n//2):
        if n%i==0:
            return "n"
    return "p"
primenumber1(5)'''
            
```
