```python
#Function to find list of factors for a given number

def factor(num):
    for i in range(1,num+1):
        if num%i==0:
            print(i)
            
factor(12)
```

```python
#Function to find the factorial of a given number

def factorial(num):
    fact=1
    for i in range(1,num+1):
        fact=fact*i
    print(fact)
    
factorial(5)
```

```python
#Function to check prime number or not

def prime(num):
    count=0
    for a in range(1,num+1):
        if num%a==0:
            count=count+1
            
    if count==2:
        print("Given number is prime number")
    else:
        print("Given number is not a prime number")
        
prime(25)
```

```python
# Function to calculate the average of first N prime number

def avgprime(N):
    count=0
    c=1
    add=1
    for r in range(1,N+1):
        for s in range(1,r+1):
            if r%s==0:
                count=count+1
        if count==2:
            add=add+r
            print("add=",add)
            c=c+1
            print("c=",c)
        avg=add/c
    print("avg=",avg) 

avgprime(6)
    
```
