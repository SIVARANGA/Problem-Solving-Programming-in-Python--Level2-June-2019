```python
# Validating the mobile number

import re
def validatemobilenumber(number):
    pattern='^[6-9][0-9]{9}$|^[0][6-9][0-9]{9}|^[+][9][1][6-9][0-9]{9}$'
    
    if re.match(pattern,str(number)):
        return True
    return False
        
        
validatemobilenumber('9640494680')

```

```python
import re
def validatemailid(email):
   # pattern='^[0-9a-z][0-9a-z_.]{4,13}[0-9a-z][@][a-z0-9]{3,18}[.][a-z][2,4]$'
     
    pattern= "^[0-9a-z][0-9a-z_.]{4,13}[0-9a-z][@][a-z0-9]{3,18}[.][a-z]{2,4}$"
    if re.match(pattern,email):
        return True
    return False


validatemailid('jbxsjhjj@gmail.com')
    
```

```python
def emailValidator(email):
    pattern = "^[0-9a-z][0-9a-z_.]{4,13}[0-9a-z][@][a-z0-9]{3,18}[.][a-z]{2,4}$"
    if re.match(pattern, email):
        return True
    return False

emailValidator("jhgdh@hgjd.com")
```

```python
import re

def phoneNumberValidator(number):
    pattern = '^[6-9][0-9]{9}$|^[0][6-9][0-9]{9}|^[+][9][1][6-9][0-9]{9}$'
    if re.match(pattern, str(number)):
        return True
    return False


phoneNumberValidator('9640464880')
```

```python
contacts={"name1":{9640494680,"ranga@gmail.com"},"name2":{9494594622,"gdsjhg@hcd.com"}}

def addcontacts(name,phonenumber,mailid):
    if name in contacts:
        print("name already exist")
        
    else:
        if not validatemobilenumber(phonenumber):
            print("invalid phone number")
            
        if not validatemailid(mailid):
            print("invalid mail id ")
            
            
        newcontacts=[]
        newcontacts.append(phonenumber)
        newcontacts.append(mailid)
        contacts[name]=newcontacts
        print(name,"added succesfully")
        
    return 

addcontacts("name3",9637418523,"jhgdh@hgj.cokm")

            
```
