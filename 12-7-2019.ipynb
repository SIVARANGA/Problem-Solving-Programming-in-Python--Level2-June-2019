```python
for i in range(int(input())):
    li = input().split()
    if len(li) == 1:
        print("Valid")
    else:
        flag = True
        for i in range(1,len(li)):
            if li[i][0] != li[i-1][-1]:
                flag = False
        for i in li:
            if li.count(i)>1:
                flag = False
    if flag == True:
        print('Valid')
    else:
        print('Invalid')
```
